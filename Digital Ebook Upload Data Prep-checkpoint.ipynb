# Setup and Inputs

```python
import pandas as pd
import numpy as np
```

```python
the_encoding_style = 'cp850'
```

```python
print('To expedite the process, please fill in any missing eISBN in the manipulated file.')
print('Afterwards, please check if the ISBN are in the database.')
```

```python
month = input('Enter the three letters of the month, such as December is "Dec": ')
fiscal_year = input('Enter FY and the last two numbers, such as "FY20": ')
excelsheet = input('Parameter Excel Sheet Path: ')
rootfile = input('Root CSV-Data Path: ')
```

## Set up conditions

```python
dflist = pd.read_excel(excelsheet, sheet_name = None,encoding=the_encoding_style)
dfxlsxdetails = pd.ExcelFile(excelsheet)
rootdf = pd.read_csv(rootfile,dtype={'eISBN13':str},encoding=the_encoding_style) # Trailing 0's to be replaced
df = dflist
sl = dfxlsxdetails.sheet_names #sheet names list
# print(sl) # Legacy code to check
```

Assign Dataframes

```python
df1,df2,df3,df4,df5,df6,df7,df8,df9 = df[sl[0]],df[sl[1]],df[sl[2]],df[sl[3]],df[sl[4]],df[sl[5]],df[sl[6]],df[sl[7]],df[sl[8]]
```

```python
df1 = df1.drop_duplicates(keep='first')
df2 = df2.drop_duplicates(keep='first')
df3 = df3.drop_duplicates(keep='first')
df4 = df4.drop_duplicates(keep='first')
df5 = df5.drop_duplicates(keep='first')
df6 = df6.drop_duplicates(subset='ISBN', keep='first')
df7 = df7.drop_duplicates(subset='ISBN', keep='first')
df8 = df8.drop_duplicates(subset='ISBN', keep='first')
df9 = df9.drop_duplicates(subset='ISBN', keep='first')
```

Root dataframe and operating dataframe split

```python
rootdf = rootdf.rename(columns={'Currency':'Territory','Partners':'Partner','Units':'Net Units','Revenue':'Net Rev'})
rootdf['eISBN13'] = rootdf['eISBN13'].str.replace('\.0','')
fulldf = rootdf
```

# Full dataframe manipulation

Devices

```python
fulldf = fulldf.merge(df1[['Conversion Devices','Device']],how='left',left_on='Partner',right_on='Conversion Devices')
fulldf[['Device']] = fulldf[['Device']].fillna('Unknown')
fulldf = fulldf.drop(columns='Conversion Devices')
```

Territory

```python
fulldf = fulldf.merge(df2[['Territory','Terry Conversion']],how='left',left_on='Territory',right_on='Territory')
fulldf['Territory'] = fulldf['Terry Conversion']
fulldf = fulldf.drop(columns='Terry Conversion')
```

Partner

```python
fulldf = fulldf.merge(df3[['Conversion Table','Partners']],how='left',left_on='Partner',right_on='Conversion Table')
fulldf['Partner'] = fulldf['Partners']
fulldf = fulldf.drop(columns={'Conversion Table','Partners'})
```

### Gross Rev and Partner Share

Percentage

```python
fulldf = fulldf.merge(df4[['Calculation Table per gross','Percentage']],how='left',left_on='Partner',right_on='Calculation Table per gross')
fulldf = fulldf.drop(columns='Calculation Table per gross')
```

```python
fulldf.loc[((fulldf['Territory'] == 'JPN') |(fulldf['Territory'] == 'IND'))  & (fulldf['Partner'] == 'KindleDirect'), 'Percentage'] = .35
```

```python
fulldf['Gross Rev'] = round(fulldf['Net Rev']/fulldf['Percentage'],2)
fulldf['Partner Share'] = fulldf['Gross Rev'] - fulldf['Net Rev']
fulldf = fulldf.drop(columns='Percentage')
```

Replace KindleDirect with Kindle and drop Series

```python
fulldf.loc[(fulldf['Partner'] == 'KindleDirect'), 'Partner'] = 'Kindle'
fulldf.rename(columns = {'Series':'Series Reference'}, inplace = True) 
```

### Dataframe prior volatile merge dependencies from database

```python
basedf = fulldf #Checkpoint
fulldf
```

## Database Data

Format

```python
basedf = basedf.merge(df7[['ISBN','Format']],how='left',left_on='eISBN13',right_on='ISBN')
basedf = basedf.drop(columns='ISBN')
```

Titles

```python
basedf = basedf.merge(df7[['ISBN','Titles']],how='left',left_on='eISBN13',right_on='ISBN')
basedf = basedf.drop(columns='ISBN')
```

Licensor

```python
basedf = basedf.merge(df8[['ISBN','Licensor']],how='left',left_on='eISBN13',right_on='ISBN')
basedf = basedf.drop(columns='ISBN')
```

Imprint

```python
basedf = basedf.merge(df9[['ISBN','Imprint']],how='left',left_on='eISBN13',right_on='ISBN')
basedf = basedf.drop(columns='ISBN')
basedf
```

#### Series

Modification to DF6

```python
df6temp = df6[['ISBN','Titles']]
df6temp['ISBN'] = df6['ISBN'].str.strip('    ')
df6temp.loc[df6temp['Titles'].str.contains('     '), 'Titles'] = np.NaN
df6temp.fillna(method='ffill', axis=0, inplace=True)
df6temp.rename(columns = {'Titles':'Series'}, inplace = True) 
#df6temp #check the printout
```

Merge

```python
basedf = basedf.merge(df6temp[['ISBN','Series']],how='left',left_on='eISBN13',right_on='ISBN')
basedf = basedf.drop(columns='ISBN')
```

Dataframe prior final manipulation

```python
finaldf = basedf
basedf
```

## Final Procedure

Add in 
Month, Year, Web ID, Actual

```python
finaldf['Month'] = month
finaldf['Year'] = fiscal_year
finaldf['Web ID'] = ''
finaldf['Actual'] = 'Actual'
```

```python
finaldf
```

Rename Column and Sort Names

```python
finaldf.rename(columns = {'eISBN13':'ISBN','Titles':'Title',}, inplace = True) 
finaldf = finaldf[
    ['Month',
     'Year',
     'Web ID',
     'ISBN',
     'Series',
     'Title',
     'Format',
     'Licensor',
     'Imprint',
     'Actual',
     'Partner',
     'Device',
     'Territory',
     'Net Units',
     'Gross Rev',
     'Partner Share',
     'Net Rev',
     'Series Reference',
     'Volume']
]
```

## Final Dataframe

```python
finaldf = finaldf.drop_duplicates()
finaldf
```
